# Nuedae Outfit Generator - Prototype

A Python prototype of the core outfit generator logic for Nuedae, a fashion app that builds outfits from what you already own.

## What it does
-Reads closet data from a CSV file

-Filters items by category and minimum rating

-Assembles a complete outfit suggestion from real closet data

## Tools Used
-Python

-pandas

-Jupyter Notebook

## Status
Active develooment - Part of the Nuedae app project

```python
import pandas as pd

df= pd.read_csv("Maya_closet - Sheet1.csv")
print(df.head())
```

```python
#Show only outfits rated 5

top_rated = df[df["Rating"] == 5]
print(top_rated)
```

```python
#show only jackets

jackets = df[df["Category"] == "Jackets"]
print(jackets)
```

```python
#show only jackets rated 5 or above

top_jackets = df[(df["Category"] == "Jackets") & (df["Rating"] >=4)]
print(top_jackets)
```

```python
#Building a function that takes a category and returns top rated items
#Outfit generator thinking

def get_top_items(df, category, min_rating=4):
    filtered = df[(df["Category"] == category) & (df["Rating"] >= min_rating)]
    return filtered

tops = get_top_items(df, "Tops")
print(tops)
```

```python
#Now calling that function for a top, a bottom, and a jacket
#Building a complete outfit suggestions pulled from our real closet

def build_outfit(df):
    top = get_top_items(df, "Tops")
    bottom = get_top_items(df, "Skirts")
    jacket = get_top_items(df, "Jackets")
    
    print("Today's outfit for Maya:")
    print(f"Top: {top.iloc[0]['Item Name']}")
    print(f"Bottom: {bottom.iloc[0]['Item Name']}")
    print(f"Jacket: {jacket.iloc[0]['Item Name']}")
    
build_outfit(df)
```
