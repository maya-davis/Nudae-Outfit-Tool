# Nuedae Data Visualization

Visual analysis of closet data for the Nuedae fashion app project.
Charts explore item distribution, rating patterns, and category insights to inform the outfit generator algorithm.

## Charts
- Items per category (count plot)
- Average rating by category (bar plot)
- Rating distribution (histogram)

## Tools Used
- Python
- pandas
- seaborn
- matplotlib
- Juptyer Notebook

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "item_name": ["Off the Shoulder Top", "Mini Skirt", "Fur Coat", "Leather Pants", "Jean Jacket", "Flip Flops", "Blazer", "Maxi Dress"],
    "category": ["Tops", "Skirts", "Jackets", "Pants", "Jackets", "Shoes", "Jackets", "Dresses"],
    "color": ["Orange", "Black", "Leopard", "Black", "Denim", "Grey", "Black", "White"],
    "rating": [5, 4, 5, 5, 5, 3, 4, 5]
}

df = pd.DataFrame(data)
```

```python
print(df)
```

```python
#How many items in categories

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="category", palette=nuedae_palette)
plt.title("Items per Category in Ciara's Closet", color="white", fontsize=14, pad=15)
plt.xlabel("Category", color="white")
plt.ylabel("Number of Items", color="white")
plt.xticks(rotation=45, color="white")
plt.yticks(color="white")
plt.gca().set_facecolor("#0A0A0A")
plt.gcf().set_facecolor("#0A0A0A")
plt.gca().spines['bottom'].set_color('#333333')
plt.gca().spines['left'].set_color('#333333')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
#Average rating by category

plt.figure(figsize=(8, 5))
avg_rating = df.groupby("category")["rating"].mean().sort_values(ascending=False)
sns.barplot(x=avg_rating.index, y=avg_rating.values, palette=nuedae_palette)
plt.title("Average Rating by Category", color="white", fontsize=14, pad=15)
plt.xlabel("Category", color="white")
plt.ylabel("Average Rating", color="white")
plt.xticks(rotation=45, color="white")
plt.yticks(color="white")
plt.gca().set_facecolor("#0A0A0A")
plt.gcf().set_facecolor("#0A0A0A")
plt.gca().spines['bottom'].set_color('#333333')
plt.gca().spines['left'].set_color('#333333')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
#Distribution of ratings across the whole closet with a histogram

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="rating", bins=5, color="#D4AF37")
plt.title("Distribution of Ratings in Ciara's Closet", color="white", fontsize=14, pad=15)
plt.xlabel("Rating", color="white")
plt.ylabel("Count", color="white")
plt.xticks(color="white")
plt.yticks(color="white")
plt.gca().set_facecolor("#0A0A0A")
plt.gcf().set_facecolor("#0A0A0A")
plt.gca().spines['bottom'].set_color('#333333')
plt.gca().spines['left'].set_color('#333333')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```
