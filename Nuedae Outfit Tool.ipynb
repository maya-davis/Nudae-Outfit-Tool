# Nuedae Outfit Tool

A Python tool that stores, logs, and exports outfit data to Excel.
Built as part of the Nuedae fashion app project.

**Features:**
- Stores outfits as structured data (name, category, rating)
- Loops through all outfits and prints a summary
- Handles missing fields gracefully without crashing
- Exports the full outfit list to Excel using pandas

```python
outfits = [
    {"name": "Button Up with Pants", "category": "Work", "rating" : "4"},
    {"name": "Flowy dress", "category": "Casual", "rating": "5"},
    {"name" : "Maxi Skirt with cute top", "category": "Night"},
    {"name" : "Sweatset", "category" : "Casual", "rating" : "3"}
]

for outfit in outfits:
    try:
        print(f"{outfit['name']} - Category: {outfit['category']}- rating: {outfit['rating']}")
    except KeyError as e:
        print(f"{outfit['name']} - Missing field: {e}, skipping")
```

```python
import pandas as pd

def exportOutfits(outfits):

    df = pd.DataFrame(outfits)

    filename = f"Nuedae_Outfits.xlsx"
    df.to_excel(filename, index= False)
    
    print(f"Report Saved: {filename}")

    
```

```python
exportOutfits(outfits)
```
